```python
import sys, os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt   # ðŸ‘ˆ this is the key

sys.path.append(os.path.abspath(".."))  # so we can import from ../src
from src.utils import clean_column_names

pd.set_option("display.max_colwidth", 120)
```

```python
RAW_PATH = "../data/raw/ecommerce.csv"  
df = pd.read_csv(RAW_PATH, encoding="ISO-8859-1")
df = clean_column_names(df)
df.head(3)
```

```python
print("Shape:", df.shape)
display(df.sample(5, random_state=1))
df.info()
```

```python
# Expected columns (after clean_column_names):
# invoice_no, stock_code, description, quantity, invoice_date, unit_price, customer_id, country

# Datetime
if df["invoicedate"].dtype == "object":
    df["invoicedate"] = pd.to_datetime(df["invoicedate"], errors="coerce")

# Numeric sanity
num_cols = ["quantity", "unitprice", "customerid"]
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# Derived fields
df["line_total"] = df["quantity"] * df["unitprice"]
df["date"] = df["invoicedate"].dt.date
df["year_month"] = df["invoicedate"].dt.to_period("M").astype(str)

df[["invoicedate", "line_total"]].head()
```

```python
date_min, date_max = df["invoicedate"].min(), df["invoicedate"].max()
n_customers = df["customerid"].nunique(dropna=True)
n_products = df["stockcode"].nunique(dropna=True)
date_min, date_max, n_customers, n_products
```

```python
miss = (
    df.isna().sum()
    .sort_values(ascending=False)
    .to_frame("missing")
    .assign(pct=lambda x: (x["missing"] / len(df)).round(4))
)
miss
```

```python
dup_count = df.duplicated().sum()
dup_count
```

```python
for col in ["quantity", "unitprice", "line_total"]:
    if col in df.columns:
        plt.figure()
        df[col].dropna().hist(bins=50)
        plt.title(f"Distribution of {col}")
        plt.xlabel(col); plt.ylabel("Count")
        plt.show()
```

```python
neg = df[df["quantity"] < 0]
neg.shape, neg.head(3)
```

```python
by_country = (
    df.groupby("country")
      .agg(orders=("invoiceno", "nunique"),
           customers=("customerid", "nunique"),
           revenue=("line_total", "sum"))
      .sort_values("revenue", ascending=False)
)
by_country.head(10)
```

```python
#Monthly Trend
monthly = df.groupby("year_month").agg(revenue=("line_total","sum"), orders=("invoiceno","nunique"))
display(monthly.tail())

plt.figure()
monthly["revenue"].plot()
plt.title("Monthly revenue")
plt.xlabel("Year-Month"); plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Top products by revenue/ by count
monthly = df.groupby("year_month").agg(revenue=("line_total","sum"), orders=("invoiceno","nunique"))
display(monthly.tail())

plt.figure()
monthly["revenue"].plot()
plt.title("Monthly revenue")
plt.xlabel("Year-Month"); plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Customer review
cust = (
    df.dropna(subset=["customerid"])
      .groupby("customerid")
      .agg(orders=("invoiceno","nunique"),
           revenue=("line_total","sum"),
           first_purchase=("invoicedate","min"),
           last_purchase=("invoicedate","max"))
)
cust["days_since_last"] = (df["invoicedate"].max() - cust["last_purchase"]).dt.days
cust.sort_values("revenue", ascending=False).head(10)
```

```python
OUT = "../data/processed/working.csv"
os.makedirs(os.path.dirname(OUT), exist_ok=True)
cols = ["invoiceno","stockcode","description","quantity","invoicedate","unitprice","customerid","country","line_total"]
df[cols].to_csv(OUT, index=False)
OUT
```
